```python
# Load dataset
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("../reports/final_business_dataset.xlsx")
print(df.shape)
df.head()
```

```python
conn = sqlite3.connect("financial.db")

df.to_sql(
    "financial_data",
    conn,
    if_exists="replace",
    index=False
)

print("Table created successfully")
```

```python
#Total Sales
query = """
SELECT SUM(sales) AS total_sales
FROM financial_data
"""
pd.read_sql(query, conn)
```

```python
#Total Profit
query = """
SELECT SUM(net_profit) AS total_profit
FROM financial_data
"""

pd.read_sql(query, conn)
```

```python
#Total Assets
query = """
SELECT SUM(total_assets) AS total_assets
FROM financial_data
"""

pd.read_sql(query, conn)
```

```python
#Total Borrowings
query = """
SELECT SUM(borrowings) AS total_borrowings
FROM financial_data
"""

pd.read_sql(query, conn)
```

```python
#Top 10 Companies by Profit
query = """
SELECT company_id,
       SUM(net_profit) AS total_profit
FROM financial_data
GROUP BY company_id
ORDER BY total_profit DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
#Top 10 Companies by Sales
query = """
SELECT company_id,
       SUM(sales) AS total_sales
FROM financial_data
GROUP BY company_id
ORDER BY total_sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
#Top 10 Companies by Assets
query = """
SELECT company_id,
       MAX(total_assets) AS assets
FROM financial_data
GROUP BY company_id
ORDER BY assets DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
#Highest Borrowings
query = """
SELECT company_id,
       MAX(borrowings) AS borrowings
FROM financial_data
GROUP BY company_id
ORDER BY borrowings DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
#Average Debt Equity Ratio
query = """
SELECT AVG(debt_equity) AS avg_debt_equity
FROM financial_data
"""

pd.read_sql(query, conn)
```

```python
#Average Profit Margin
query = """
SELECT AVG(net_profit_margin) AS avg_profit_margin
FROM financial_data
"""

pd.read_sql(query, conn)
```

```python
#Top 10 Companies by Free Cash Flow
query = """
SELECT company_id,
       SUM(free_cash_flow) AS free_cash_flow
FROM financial_data
GROUP BY company_id
ORDER BY free_cash_flow DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
#Year-wise Sales Trend
query = """
SELECT year,
       SUM(sales) AS total_sales
FROM financial_data
GROUP BY year
ORDER BY year
"""

pd.read_sql(query, conn)
```

```python
#Year-wise Profit Trend
query = """
SELECT year,
       SUM(net_profit) AS total_profit
FROM financial_data
GROUP BY year
ORDER BY year
"""

pd.read_sql(query, conn)
```

```python
#Companies with Positive Free Cash Flow
query = """
SELECT company_id,
       COUNT(*) AS records
FROM financial_data
WHERE free_cash_flow > 0
GROUP BY company_id
ORDER BY records DESC
"""

pd.read_sql(query, conn)
```

```python
#Close Connection
conn.close()
print("Database connection closed")
```
