```python
import pandas as pd

master = pd.read_excel(
    "../reports/master_financial_data.xlsx"
)

master.head()
```

```python
#Top 10 Companies by Sales
top_sales = (
    master.groupby("company_id")["sales"]
    .max()
    .sort_values(ascending=False)
    .head(10)
)

print(top_sales)
```

```python
#Top 10 Companies by Net Profit
top_profit = (
    master.groupby("company_id")["net_profit"]
    .max()
    .sort_values(ascending=False)
    .head(10)
)

print(top_profit)
```

```python
#Companies with highest Debt
top_debt=(
    master.groupby("company_id")["borrowings"]
    .max()
    .sort_values(ascending=False)
    .head(10)
)
print(top_debt)
```

```python
#Create Debt-to-Equity Ratio
master["debt_equity"] = (
    master["borrowings"] /
    (
        master["equity_capital"] +
        master["reserves"]
    )
)
```

```python
#Highest Debt-to-Equity Companies
risk = (
    master.groupby("company_id")["debt_equity"]
    .max()
    .sort_values(ascending=False)
    .head(10)
)

print(risk)
```

```python
#Net Profit Margin
master["net_profit_margin"] = (
    master["net_profit"] /
    master["sales"]
) * 100
```

```python
#Most Profitable Companies
margin = (
    master.groupby("company_id")["net_profit_margin"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(margin)
```

```python
#Free Cash Flow
master["free_cash_flow"] = (
    master["operating_activity"] +
    master["investing_activity"]
)
```

```python
#Best Cash Generating Companies
cash = (
    master.groupby("company_id")["free_cash_flow"]
    .max()
    .sort_values(ascending=False)
    .head(10)
)

print(cash)
```

```python
#Save Final Dataset
master.to_excel(
    "../reports/final_business_dataset.xlsx",
    index=False
)

print("Business dataset saved")
```
