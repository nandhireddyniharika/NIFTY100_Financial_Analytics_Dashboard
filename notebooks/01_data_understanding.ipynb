```python
#Complete Summary Table
summary = []

for name, df in datasets.items():

    summary.append([
        name,
        df.shape[0],
        df.shape[1],
        df.isnull().sum().sum(),
        df.duplicated().sum()
    ])

summary_df = pd.DataFrame(
    summary,
    columns=[
        "Table",
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicates"
    ]
)

summary_df
```

```python
#Save Data Understanding Report
summary_df.to_excel(
    "../reports/data_summary.xlsx",
    index=False
)

print("Data summary saved")
```
