```python
import pandas as pd
import numpy as np
```

```python
analysis = pd.read_excel("../data/raw/analysis.xlsx")
balancesheet = pd.read_excel("../data/raw/balancesheet.xlsx")
cashflow = pd.read_excel("../data/raw/cashflow.xlsx")
companies = pd.read_excel("../data/raw/companies.xlsx")
documents = pd.read_excel("../data/raw/documents.xlsx")
profitandloss = pd.read_excel("../data/raw/profitandloss.xlsx")
prosandcons = pd.read_excel("../data/raw/prosandcons.xlsx")

datasets = {
    "analysis": analysis,
    "balancesheet": balancesheet,
    "cashflow": cashflow,
    "companies": companies,
    "documents": documents,
    "profitandloss": profitandloss,
    "prosandcons": prosandcons
}
```

```python
clean_data = {}

for name, df in datasets.items():

    # First row contains actual column names
    df.columns = df.iloc[0]

    # Remove first row
    df = df[1:].reset_index(drop=True)

    clean_data[name] = df

print("Headers fixed successfully")
```

```python
for name, df in clean_data.items():

    # Remove leading/trailing spaces
    df = df.map(
        lambda x: x.strip()
        if isinstance(x, str)
        else x
    )

    # Convert blank strings to NaN
    df.replace(
        r'^\s*$',
        np.nan,
        regex=True,
        inplace=True
    )

    # Convert text nulls to NaN
    df.replace(
        ["null", "NULL", "Null"],
        np.nan,
        inplace=True
    )
    clean_data[name] = df

print("Null and blank values cleaned")
```

```python
for name, df in clean_data.items():

    # Remove rows where all columns are empty
    df.dropna(
        how="all",
        inplace=True
    )

    clean_data[name] = df

print("Empty rows removed")
```

```python
for name, df in clean_data.items():

    df.drop_duplicates(
        inplace=True
    )

    clean_data[name] = df

print("Duplicate rows removed")
```

```python
for name, df in clean_data.items():

    print("\n")
    print("="*50)
    print(name.upper())
    print("="*50)

    missing = df.isnull().sum()

    missing = missing[missing > 0]

    if len(missing) == 0:
        print("No missing values")

    else:
        print(missing)
```

```python
summary = []

for name, df in clean_data.items():

    summary.append([
        name,
        df.shape[0],
        df.shape[1],
        df.isnull().sum().sum(),
        df.duplicated().sum()
    ])

clean_summary = pd.DataFrame(
    summary,
    columns=[
        "Table",
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicates"
    ]
)

clean_summary
```

```python
for name, df in clean_data.items():

    df.to_excel(
        f"../data/clean/{name}_clean.xlsx",
        index=False
    )

print("All cleaned files saved successfully")
```

```python
clean_summary.to_excel(
    "../reports/data_cleaning_summary.xlsx",
    index=False
)

print("Cleaning report saved")
```
